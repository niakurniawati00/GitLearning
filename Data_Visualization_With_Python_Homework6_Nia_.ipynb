```python
import pandas as pd

file_path = 'credit-card-customers/BankChurners.csv'
data = pd.read_csv(file_path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Tugas nomor 1

```python
##1. Pie chart Attrition Flag
attrition_counts = data['Attrition_Flag'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66B3FF', '#FF9999'])
plt.title('Proporsi Attrition Flag')
plt.show()
```

```python
##2. Vertical bar chart untuk customer_age
plt.figure(figsize=(8, 6))

sns.histplot(data['Customer_Age'], kde=False, bins=20, color='skyblue')
plt.title('Distribusi Umur Pelanggan')
plt.xlabel('Usia')
plt.ylabel('Jumlah Pelanggan')
plt.show()
```

```python
##3. Donut chart untuk gender
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['#FF9999','#66B3FF'])
# Membuat donut
circle = plt.Circle((0,0), 0.70, color='white')
plt.gca().add_artist(circle)
plt.title('Distribusi Gender Pelanggan')
plt.show()
```

```python
##4. Horizontal bar chart untuk education_level
plt.figure(figsize=(8, 6))
sns.countplot(y='Education_Level', data=data, hue='Education_Level', palette='pastel', legend=False)
plt.title('Distribusi Level Pendidikan Pelanggan')
plt.xlabel('Jumlah Pelanggan')
plt.ylabel('Tingkat Pendidikan')
plt.show()
plt.show()
```

```python
##5. Vertical bar chart untuk income_category
plt.figure(figsize=(8, 6))
sns.countplot(x='Income_Category', data=data, hue="Income_Category", palette='pastel', legend=False)
plt.title('Distribusi Kategori Penghasilan Pelanggan')
plt.xlabel('Kategori Penghasilan')
plt.ylabel('Jumlah Pelanggan')
plt.show()
```

#Tugas nomor 2

1. Pie Chart (attrition_flag): Diagram ini menggambarkan perbandingan antara pelanggan yang berhenti dan yang tetap menggunakan layanan.
Insight: Jika sebagian besar pelanggan tetap bertahan, perusahaan dapat fokus menjaga hubungan dengan mereka dan meningkatkan program loyalitas untuk mencegah kehilangan pelanggan.

2. Diagram Batang Vertikal (customer_age): Menunjukkan distribusi usia pelanggan dengan jumlah pelanggan terbanyak di rentang usia tertentu.
Insight: Pelanggan yang lebih muda atau dewasa cenderung lebih aktif menggunakan kartu kredit, menjadikannya segmen yang potensial untuk produk atau promosi baru.

3. Diagram Donat (gender): Perbandingan antara jumlah pelanggan pria dan wanita.
Insight: Jika pelanggan dari jenis kelamin tertentu lebih banyak, strategi pemasaran bisa lebih difokuskan pada preferensi dan kebiasaan belanja mereka.

4. Diagram Batang Horizontal (education_level): Menunjukkan sebaran tingkat pendidikan pelanggan.
Insight: Pelanggan dengan tingkat pendidikan lebih tinggi mungkin memiliki daya beli yang lebih besar dan lebih cenderung menggunakan layanan kredit. Mereka bisa menjadi target yang baik untuk produk premium.

5. Diagram Batang Vertikal (income_category): Menunjukkan distribusi penghasilan pelanggan.
Insight: Pelanggan dengan penghasilan lebih tinggi mungkin lebih sering melakukan transaksi atau menggunakan kartu kredit. Oleh karena itu, mereka bisa menjadi fokus untuk promosi produk dengan manfaat yang lebih besar.

#Tugas nomor 3

```python
# Countplot untuk gender dan attrition_flag
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='Attrition_Flag', data=data, palette='coolwarm')
plt.title('Distribusi Gender Berdasarkan Attrition Flag')
plt.xlabel('Gender')
plt.ylabel('Jumlah Pelanggan')
plt.show()
```

#Tugas nomor 4


```python
# Scatter plot Total_Trans_Amt vs Total_Trans_Ct, dipisah berdasarkan Marital_Status
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Total_Trans_Amt', y='Total_Trans_Ct', hue='Marital_Status', palette='tab10', alpha=0.7)

# Menambahkan label dan judul
plt.title('Total Transaction Amount vs Total Transaction Count by Marital Status', fontsize=16)
plt.xlabel('Total Transaction Amount', fontsize=12)
plt.ylabel('Total Transaction Count', fontsize=12)
plt.legend(title='Marital Status')

# Menampilkan plot
plt.show()
```

#Tugas nomor 5

```python
# kolom yang ingin disertakan dalam heatmap
selected_cols = ['Credit_Limit', 'Total_Revolving_Bal', 'Avg_Open_To_Buy', 'Avg_Utilization_Ratio']

# Hitung matriks korelasi untuk kolom yang dipilih
correlation_matrix = data[selected_cols].corr()

# Buat heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi untuk Kolom yang Dipilih')
plt.show()
```
