#Menghilangkan Missing Values Data

#Tugas No 1

##1.1 Menambhan Nama Kolom "Full Name"

```python
# Menggabungkan 'First Name' dan 'Last Name' menjadi 'Full Name' dengan format title case
df1['Full Name'] = (df1['First Name'] + ' ' + df1['Last Name']).str.title()
```

##1.2 Memindahan Posisi Kolom

```python
cols = list(df1.columns)
last_name_index = cols.index('Last Name')
cols.insert(last_name_index + 1, cols.pop(cols.index('Full Name')))
df1 = df1[cols]
```

##Hasil

```python
df1.head(2)
```

#Tugas No.2

##2.1 Kolom GMV

```python

df1.loc[:, 'GMV'] = df1['Transaction Amount'] - df1['Seller Discount'] + df1['Delivery Fee']
```

##2.2 Memindahkan posisi kolom

```python
cols = list(df1.columns)
eller_discount_index = cols.index('Seller Discount')
cols.insert(eller_discount_index + 1, cols.pop(cols.index('GMV')))
df1 = df1[cols]
```

```python
df1.head(2)
```

#Tugas 3

#3.1 Membuat group

```python
df1['Meta Category'].unique()
```

```python
def group_meta_category(category):
    # Menentukan grup berdasarkan kategori
    if category in ['Home/Food', 'Sports/Hobi/Otomotif', 'Fashion', 'Beauty/Health']:
        return 'Group 1'
    else:
        return 'Group 2'
```

```python
df1.loc[:, 'Group'] = df1['Meta Category'].apply(group_meta_category)
```

```python
df1.head(6)
```

```python
df1['Group'].value_counts()
```

##3.2 Pivot

```python
df1.loc[:, "Month"] = df1["Paid Date"].dt.to_period('M')
```

```python
pivot_df1 = df1.pivot_table(
    index='Month',       # Mengelompokkan berdasarkan bulan
    columns='Group',     # Mengelompokkan berdasarkan Group
    values='GMV',        # Mengambil nilai GMV
    aggfunc='sum',       # Menjumlahkan nilai GMV
    fill_value=0         # Mengisi nilai yang kosong dengan 0
)

# Menampilkan Pivot Table
print(pivot_df1)
```
